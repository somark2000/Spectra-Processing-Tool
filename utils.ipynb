## Add axis and labels for zoom in screenshots for spectra

* in this script you are able to insert as many screenshots as you wish to add the proper axis 
* at the moment if the images are associated with different axis values, please generate them in separate runs, otherwise update the script to handle lists for the spectra limits

```python
import matplotlib.pyplot as plt
import numpy as np

# img file location
files = ["C:/Users/maria/OneDrive - Universitatea Babe-Bolyai/Desktop/dna teo/zoom in_fig5.png","C:/Users/maria/OneDrive - Universitatea Babe-Bolyai/Desktop/dna teo/zoom in_fig4.png"]
lower_end_xLabel = 612
upper_end_xLabel = 687
lower_end_yLabel = 0.55
upper_end_yLabel = 1.01
step_size_xLabel = 10
step_size_yLabel = 0.05

for file in files:
    img = plt.imread(file)
    fig, ax = plt.subplots()

    x = np.arange(lower_end_xLabel,upper_end_xLabel,step_size_xLabel)
    y = np.arange(lower_end_yLabel,upper_end_yLabel,step_size_yLabel)

    ax.set_xlim((lower_end_xLabel,upper_end_xLabel))
    ax.set_ylim((lower_end_yLabel,upper_end_yLabel))
    ax.set_xticks(x,labels=[f"{e}" for e in x],fontsize=12)
    ax.set_yticks(y,labels=[f"{e:1.2f}" for e in y],fontsize=12)
    ax.set_xlabel("Wavelength [nm]",fontsize=17)
    ax.set_ylabel("Absorbance [a.u.]",fontsize=17)
    ax.imshow(img,extent=ax.get_xlim() + ax.get_ylim(), aspect='auto', zorder = -1)
    ax.plot()
```
