## Add axis and labels for zoom in screenshots for spectra

* in this script you are able to insert as many screenshots as you wish to add the proper axis 
* at the moment if the images are associated with different axis values, please generate them in separate runs, otherwise update the script to handle lists for the spectra limits

```python
import matplotlib.pyplot as plt
import numpy as np

# img file location
files = ["C:/Users/smark/Downloads/dls_AgNPs 2305_1.png","C:/Users/smark/Downloads/dls_AgSiO2 2305_1.png",
         "C:/Users/smark/Downloads/spectru imag AgNPs_stanga.png","C:/Users/smark/Downloads/spectru imag AgNPs_dreapta.png"]
lower_end_xLabel = [-1,-1,400,400]
upper_end_xLabel = [2,2.447158,1000.1,1000.1]
lower_end_yLabel = [0,0,-30,-20]
upper_end_yLabel = [40.01,30.01,320.1,520.1]
step_size_xLabel = [10,10,100,100]
step_size_yLabel = [5,5,50,100]
y_delta = [0,0,30,20]
x_delta = [0,0,0,0]
x_title = ["Size [d.nm]","Size [d.nm]","Wavelength [nm]","Wavelength [nm]"]
y_title = ["Number [Percent]","Number [Percent]","Value","Value"]
xlog = [True,True,False,False]
dark = [False,False,True,True]
title = [None,None,"Spectral Profile","Spectral Profile"]

for i,file in enumerate(files):
    img = plt.imread(file)
    if dark[i]:
        plt.style.use('dark_background')
    fig, ax = plt.subplots()
    if xlog[i]:
        x = np.logspace(lower_end_xLabel[i],upper_end_xLabel[i],step_size_xLabel[i])
        x=np.delete(x,[1,2,3,4])
    else:
        x = np.arange(lower_end_xLabel[i]+x_delta[i],upper_end_xLabel[i]+x_delta[i],step_size_xLabel[i])
    y = np.arange(lower_end_yLabel[i]+y_delta[i],upper_end_yLabel[i]+y_delta[i],step_size_yLabel[i])
    # print(x)

    ax.set_xlim((lower_end_xLabel[i],upper_end_xLabel[i]))
    ax.set_ylim((lower_end_yLabel[i],upper_end_yLabel[i]))
    ax.set_xticks(x,labels=[f"{e:1.0f}" for e in x],fontsize=12)
    ax.set_yticks(y,labels=[f"{e:1.0f}" for e in y],fontsize=12)
    ax.set_xlabel(x_title[i],fontsize=17)
    ax.set_ylabel(y_title[i],fontsize=17)
    ax.imshow(img,extent=ax.get_xlim() + ax.get_ylim(), aspect='auto', zorder = -1)
    if title[i] != None:
        ax.set_title(title[i])
    ax.plot()
    plt.style.use('default')
```
